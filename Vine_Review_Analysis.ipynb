```python
# Import pandas
import pandas as pd
```

```python
# Read vine_table.csv int opandas
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
# Filter dataframe to only show reviews with over 20 total votes
votes_df = vine_df.loc[vine_df['total_votes'] >= 20]
votes_df.head()
```

```python
# Filter votes_df to show reviews that were labeled as helpful at least 50% of the time
votes_50_df = votes_df.loc[votes_df['helpful_votes'] / votes_df['total_votes'] >=.5] 
votes_50_df.head()
```

```python
# Filter votes_50_df to show only vine program reviews
paid_df = votes_50_df.loc[votes_50_df['vine'] == 'Y']
paid_df.head()
```

```python
# Filter votes_50_df to show only non-vine program reviews

unpaid_df = votes_50_df[votes_50_df['vine'] == 'N']
unpaid_df.head()
```

```python
# Calculate total votes for all vine program reviews
paid_total_votes = paid_df['total_votes'].sum()
# Calculate vine program 5 star review amount
paid_5star_votes_df = paid_df[paid_df['star_rating'] == 5]
paid_5star_count = paid_5star_votes_df['star_rating'].count()
# Calculate percentage of 5 star reviews for viune program reviews
paid_5star_percentage = paid_5star_count / paid_total_votes
print(f'The total number of reviews for the vine program were {paid_total_votes}')
print(f'The number of 5 star votes for the vine program were {paid_5star_count}')
print(f'The percentage of 5 star reviews for the Vine program was {paid_5star_percentage}')
```

```python
# Calculate total votes for all vine program reviews
unpaid_total_votes = unpaid_df['total_votes'].sum()
# Calculate vine program 5 star review amount
unpaid_5star_votes_df = unpaid_df[unpaid_df['star_rating'] == 5]
unpaid_5star_count = unpaid_5star_votes_df['star_rating'].count()
# Calculate percentage of 5 star reviews for viune program reviews
unpaid_5star_percentage = unpaid_5star_count / unpaid_total_votes
print(f'The total number of reviews not using the vine program was {unpaid_total_votes}')
print(f'The number of 5 star votes not using the vine program program was {unpaid_5star_count}')
print(f'The percentage of 5 star reviews not using the vine program was {unpaid_5star_percentage}')

```
